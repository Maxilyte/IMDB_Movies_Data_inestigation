## Exploratory Data Analysis

### Question 1: Which Genre are the most popular from year to year

```python
# here i want to generate a new dataframe of genre counts in relation to the years of movie production
df_genre = df_clear.groupby(['release_year', 'genres'])['id'].count().reset_index() #I used id to count on the genres
df_genre.rename(columns = {'id': 'counts'}, inplace=True) #rename id column to 'counts'
df_genre.head()
```

```python
# now i want to look at the movie genres to know which one appeard most or have been popular over the year
genre_Popularity = df_genre.iloc[df_genre.groupby('release_year')['counts'].agg(pd.Series.idxmax)]
genre_Popularity.tail(10) # i used tail so as to get the values from the most recent years 
```

```python
popular_movies = genre_Popularity.tail(10)
popular_movies.head()
```

### Answer

> From the look at the data above, you will agree with me that the Genre"comedy" has been the most popular over the year.  so lets take a look at the visualisations to have a clear understanding of the data above

```python
popular_movies.pivot(index='release_year',columns='genres', values='counts').plot(kind='bar', figsize=(15,8)) # i had to pivot the table so as to get a clear visualization of the datas  i am working on

plt.title('Count of most produced genres year to year from 1996 to 2015') # i didnt include all the years here so that my visualization can be clearer
```

### Answer
> From the graph above we can see that comedy( represented by orage) is the most popular over the selected years with the highest count in 2015, follwoed by Drama( ther green legend); with the highest count in the year 2009 and then Action|Thriller.   


### Q2: Which of the production companies produce the most movies



> i am going to introduce a new function that i learnt newly to answer this question. its is called a 'get_dummies function. it is used to convert categorical variable into dummy/indicator variables. i am also going to use the sort function to arrange my values and finally the 'lambda function' which is also known as the annonymous function for my arguments  so lets see how it works 

```python
dp = df_clear['production_companies'].str.get_dummies(sep='|')
```

```python
dp_movies = dp[dp.columns].apply(lambda x: sum(x.values))
```

```python
plt.figure(figsize = (7,7))
dp_movies.sort_values(0,ascending=False).head(15).plot.pie(autopct="%1.1f%%",pctdistance = 1.05, labeldistance = 1.1);
plt.axis('equal');
plt.show()
```

> From the above pie chart visualization we can see that 'Universal pictures' produce the most films over the year with 14.4% followed by 'Warner Bros' with the total of 14.1% and the third is 'Columbia Pictures' with 8.9% production. while the Lengendary Pictures and DreamWorks SKG has the least production 
>

    1. Universal Picturs

### Question 4: Relationship between the bugets and the generated revenue of each year

```python
# lets look at try to select the columns we want to use once more  
# we will be looking at the two columns that contains "budget" and the "revenue" column  on each year
# we use the grouby function once again

#so lets use the groupby function
Relationships = df_clear.groupby(['director'])[['budget','revenue']].sum()


Relationships.head()
```

```python
Relationships.describe()
```

```python
#lets visualize it using seaborn and see
plt.figure(figsize=(10,5))
plt.scatter(x ='budget', y='revenue', data=Relationships,)
plt.xlabel('Revenue',color='r', fontsize=20)
plt.ylabel('Budget',color='b', fontsize=20)
plt.title('Relationship between budget and Generated Revenue',fontsize = 20)
plt.show()
```

### Question 5: What is the Average runtime of the movies in the dataset?

> We will create a little function called "Average" This can take the average of the runtine column when executed. it can also take the average of anyother columns, lets says "revenue" or budget etc. All we need is to call in the column that we want to analyse

```python
# defining a function to find average of a column
def Average(column):
    return df_clear[column].mean()
```

```python
#calling above function
Average('runtime')
```

> From the result above, we can see that the average runtime of the movies in the whole dataset is approximately 113

```python
#Finally let me use a histogram to plot this visualisation

# inputing the shape of the graph
plt.figure(figsize=(9,5), dpi = 100)
plt.xlabel('Runtime of the Movies',color='r', fontsize= 20)
plt.ylabel('Nos.of Movies in the Dataset',color='y', fontsize=20)
plt.title('Runtime of all the movies', fontsize=20)
plt.hist(df_clear['runtime'], rwidth = 0.9, bins =35)

plt.show()
```
